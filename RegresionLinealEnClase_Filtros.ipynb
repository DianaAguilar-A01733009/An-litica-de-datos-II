```python
# Declarar las variables dependientes e independientes para la regresin lineal
Vars_Indep1 = df[['alcohol']]
Vars_Dep = df['total']
```

```python
#Se define model como la funcin de regresin lineal
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
#Para ver que si me lo est guardando como modelo de sklearn
type(model1)
```

```python
model1.fit(X = Vars_Indep1, y = Vars_Dep)
```

```python
model1.__dict__
```

```python
#El resultado es el coeficiente de correlacin
model1.score(Vars_Indep1, Vars_Dep)
```
